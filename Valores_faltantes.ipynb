## Trabajaremos con los datos nulos o faltantes 

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Data_new_table.csv",delimiter=";")              # SET DE DATOS QUE NO SUFRIRA CAMBIOS 
data_complem = pd.read_csv("Data_new_table.csv",delimiter=";")      # SET DE DATOS PARA MODIFICAR 
```

```python
def recorte_nulos(data):
    data = pd.Series(data)
    cont = 0    # contamos los datos diferentes de 0
    for value in data:
        if value != 0.0:
            cont +=1
    media_datos_existentes = data.sum()/cont
    
    return media_datos_existentes
```

```python
# Variables para filtrado
PBI = "GB.XPD.RSDV.GD.ZS"
paises = data['Country Code'].unique()

# Agregamos una funcion para calcular la media recortada
def media_recortada(data):
    trim_percentage = 0.1       # media recortada al 10%
    sorted_data = np.sort(data)
    trim_size = int(len(data) * trim_percentage / 2)
    
    # Recortamos los valores más bajos y más altos
    trimmed_data = sorted_data[trim_size : -trim_size]
    
    # Calculamos la media de los valores restantes
    mean = np.mean(trimmed_data)
    return mean
```

```python
marcador_pbi = [] # 1: pais con datos de pbi; 0: pais con pocos o nulos datos de pbi

for pais in paises:    
    # datos filtrados
    data_fil = data[(data['Country Code'] == pais) &  (data['Indicator Code'] == "GB.XPD.RSDV.GD.ZS")]
    media_rect = media_recortada(data_fil['Valor'].values)
    if media_rect == 0:
        marcador_pbi.append(0)
    else:
        marcador_pbi.append(1)
```

```python
# Filtramos los paises que cuentan con datos relevantes de pbi
pais_con_pbi = []
for i in range(0,len(paises)):
    pais = paises[i]
    if marcador_pbi[i] == 1:
        pais_con_pbi.append(pais)
```

```python
# AGREGAMOS EL VALOR DE MEDIA A LOS PAISES QUE SI CUENTAN CON DATOS DE PBI
# y ademas tienen muestras de datos faltantes
for pais in pais_con_pbi:    
    # datos filtrados
    data_fil = data[(data['Country Code'] == pais) &  (data['Indicator Code'] == "GB.XPD.RSDV.GD.ZS")]
    media = recorte_nulos(data_fil['Valor'].values)
    for index in data_fil.index:
        if data_complem.loc[index,'Valor'] == 0.0:
            data_complem.loc[index,'Valor'] = media
```

```python
for pais in pais_con_pbi[0:3]:   
    # datos filtrados
    data_fil = data[(data['Country Code'] == pais) &  (data['Indicator Code'] == "GB.XPD.RSDV.GD.ZS")]
    print(data_fil)
```

```python
# Comprobacion de cambio de datos 
for pais in pais_con_pbi[0:15]:    
    # datos filtrados
    data_fil = data_complem[(data_complem['Country Code'] == pais) &  (data_complem['Indicator Code'] == "GB.XPD.RSDV.GD.ZS")]
    print(data_fil)
```

# PATENTES RESIDENTES

```python
# ANALISIS INDEPENDIENTE
# filtro para patentes residentes "IP.PAT.RESD"

conteo_datos = []       # suma de años en los que se producio patentes
intermedios = []        # paises que producieron patentes en un periodo de 9 a 15 años, intercalados

# SOLO PARA ANALISIS DE DATOS 
for pais in paises:     
    pat_res = data[(data['Country Code'] == pais) & (data['Indicator Code']== "IP.PAT.RESD")]
    cont = 0        # servira para contar la cantidad de valores nulos
    for value in pat_res['Valor'].values:
        if value != 0.0:
            cont += 1
    
    conteo_datos.append(cont)       # Agregamos el contador a la lista de datos 
    
    if (cont >= 9) and (cont <= 15): # verificamos el rango para posterior analisis 
        intermedios.append(pais)
    
    
```

```python
# Solo 16 paises ingresaron a la categoria 2, se procedio a analizarlos
for pais in intermedios:     
    pat_res = data[(data['Country Code'] == pais) & (data['Indicator Code']== "IP.PAT.RESD")]
    print(pat_res)
```

```python
# 
categoria_resd = []     # categoria asignada de acuerdo a la cantidad de años que producieron patentes
                            # 0: sin datos (sin produccion de patentes)
                            # 1: años_prooducidos < 9 (baja o nula produccion)
                            # 2: 9 <= años_prooducidos <= 15 (poca produccion) [excepcion: NIGERIA Y COREA]
                            # 3: años_prooducidos > 15 (produccion constante de patentes)  

for pais in paises:     
    pat_res = data[(data['Country Code'] == pais) & (data['Indicator Code']== "IP.PAT.RESD")]
    cont = 0        # servira para contar la cantidad de valores nulos
    for value in pat_res['Valor'].values:
        if value != 0.0:
            cont += 1

    if cont == 0:                       categoria_resd.append(0) 
    elif (cont > 0) & (cont < 9):       categoria_resd.append(1)
    elif (cont >= 9) and (cont <= 15):  categoria_resd.append(2)
    else:                               categoria_resd.append(3)
    
```

```python
# SOLO PARA VISUALIZACION
cont_data = { 'pais': paises,
            'conteo':conteo_datos,
            'categoria_resd': categoria_resd}
data_contada = pd.DataFrame(cont_data)
```

```python
# Haremos una intervencion en los datos para los paises de categoria 3, se llenara los nulos con la media de los valores 
# que se tengan, se tomo esta medida ya que se cuenta con una cantidad considerable de datos

pais_cat3 = list(data_contada['pais'][data_contada['categoria_resd'] == 3]) # paises categoria 3

for pais in pais_cat3:     
    pat_res = data[(data['Country Code'] == pais) & (data['Indicator Code']== "IP.PAT.RESD")]
    media = recorte_nulos(pat_res['Valor'].values)

    for index in pat_res.index:
        if data_complem.loc[index,'Valor'] == 0.0:
            data_complem.loc[index,'Valor'] = media

```

```python
for pais in pais_cat3[0:3]:   
    # datos filtrados
    data_fil = data[(data['Country Code'] == pais) &  (data['Indicator Code'] == "IP.PAT.RESD")]
    print(data_fil)
```

```python
# Comprobacion de cambio de datos 
for pais in pais_cat3[0:3]:    
    # datos filtrados
    data_fil = data_complem[(data_complem['Country Code'] == pais) &  (data_complem['Indicator Code'] == "IP.PAT.RESD")]
    print(data_fil)
```

# CANTIDAD DE CIENTIFICOS (POR CADA MILLOS DE PERSONAS): "SP.POP.SCIE.RD.P6"

Variable relacionada a personas involucradas en la produccion de material cientificos
Se realizara un conteo de la cantidad de años que tienen registros de esta variable, ya que por coherencia tambien habra paises que tengo una produccion baja o nula de cientificos, en los casos donde se tengan registros intercalados se procedera a rellenar los valores faltantes del periodo [2000 : 2019], ya que la media de produccion en este periodo de tiempo sera usada para visualizar la produccion de cientificos del pais.

```python
cont_finales = []       # almeacena el conte de los datos de los paises en los ultimos 5 años
categoria_scie = []     # indica la presencia de datos:
                        # 0: sin datos, 1: con datos

for pais in paises:
    scie_cont = data[(data['Country Code'] == pais) & (data['Indicator Code']== "SP.POP.SCIE.RD.P6")]
    cont2 = 0       # conteo de datos en los ultimos 5 años 
    cal_med = list(scie_cont['Valor'].values)

    # -----------CATEGORIA PARA CANTIDAD DE DATOS "ACTUALIZADOS"--------------------
    for val in cal_med[-5:]:
        if val != 0.0:
            cont2 += 1

    if cont2 == 0:  categoria_scie.append(0)
    else:           categoria_scie.append(1)

    cont_finales.append(cont2)
    # -------------------------------

    if cont2 > 0:   
        media_rect_nul = recorte_nulos(cal_med) 
        for index in scie_cont.index:
            if data_complem.loc[index,'Valor'] == 0.0:
                data_complem.loc[index,'Valor'] = media_rect_nul
```

```python
# Solo para visualizacion
data_scie = {'pais': paises,
'cont_fin': cont_finales}
scie_data = pd.DataFrame(data_scie)

paises_con_datos = scie_data[scie_data.cont_fin != 0]
paises_scie = paises_con_datos.pais.values
```
