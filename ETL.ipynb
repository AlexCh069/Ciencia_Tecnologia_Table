```python
import pandas as pd 
import os
```

```python
archivo = open('API_14_DS2_es_csv_v2_4533287.csv','r') # Abrimos archivo
# Separacion de datos
data = []
for fila in archivo:
    dato = fila.split(',"')          
    data.append(dato)
archivo.close()

# Limpieza de datos 
rep = []
for i in range(0,len(data)):
    fila = data[i]
    fil = []
    for k in fila:
        dato = k.strip('"')
        fil.append(dato)
    rep.append(fil)

# Ingesta de datos a Dataframe 
data_db = pd.DataFrame(columns = rep[0])
for colum in range(0,len(data_db.columns)):
    data_db[f'{data_db.columns[colum]}'] = pd.Series([i[colum] for i in rep])

data_db.drop(0,inplace = True)  # Borramos la primera fila
data_db.drop(columns=['\n','Indicator Name'],inplace = True) # Eliminamos la ultima columna
```

```python
años = ['Country Code', 'Indicator Code','1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997','1998', '1999', 
        '2000', '2001', '2002', '2003', '2004', '2005', '2006','2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
        '2016', '2017', '2018', '2019', '2020', '2021']
        
data_db = data_db[años] # Columnas desde 1990 - 2021
data_db = data_db.replace({"": 0})      # reemplazamos valores vacios con 0

for año in range(1990,2022):
    data_db[f'{año}'] = data_db[f'{año}'].astype('float')       # Cambio de tipo de dato (float)
```

```python
country = pd.read_csv('Metadata_Country_API_14_DS2_es_csv_v2_4533287.csv')
country = country[['Country Name', 'Country Code', 'Region', 'Income_Group']]
country = country.fillna('Sin Dato')

ctr_names = []
for ctr in country['Country Name'].values:
    # ctr_name = country.loc[ctr,'Country Name']
    ctr_name = ctr.strip('\r\n')
    ctr_name = ctr_name.replace(',',".")
    ctr_names.append(ctr_name)

country['Country_Name'] = pd.Series(ctr_names)
country = country[['Country_Name', 'Country Code', 'Region', 'Income_Group']]
```
